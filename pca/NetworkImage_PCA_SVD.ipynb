```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for visualization
nodes = ['Node A', 'Node B', 'Node C', 'Node D']
connections = [(0, 1), (1, 2), (2, 3), (3, 0)]
latency = [10, 20, 15, 25]  # latency in milliseconds

# Create a network graph
plt.figure(figsize=(8, 6))
plt.title('Network Topology and Latency')

# Plot nodes
for i, node in enumerate(nodes):
    plt.scatter(i, 0, label=node, s=100)

# Plot connections
for conn in connections:
    plt.plot([conn[0], conn[1]], [0, 0], 'k-', color='blue', lw=2)

# Annotate latencies
for i, latency_value in enumerate(latency):
    plt.text(i, 0.05, f'{latency_value} ms', ha='center')

plt.xlabel('Network Nodes')
plt.ylabel('Latency')
plt.legend()
plt.grid(True)
plt.show()
```
