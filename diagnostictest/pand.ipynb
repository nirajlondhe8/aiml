```python
import unittest

import importlib
import pand

importlib.reload(pand)

from pand import Board, Solver


class BoardTest(unittest.TestCase):
    def test_move_returns_caseload_before_spread(self):
        board = Board(city_count=5, epicenter=3)

        self.assertEqual(board.move(2), 0)
        self.assertEqual(board.day, 1)
        self.assertEqual(board.infected_city_numbers(), [2, 3, 4])

    def test_epicenter_has_caseload_one_at_start(self):
        board = Board(city_count=5, epicenter=3)

        self.assertEqual(board.move(3), 1)

    def test_caseload_increases_after_city_is_infected(self):
        board = Board(city_count=5, epicenter=3)

        board.move(1)
        board.move(1)

        self.assertEqual(board.move(2), 2)

    def test_invalid_city_count_raises_error(self):
        with self.assertRaises(ValueError):
            Board(city_count=0, epicenter=1)

    def test_invalid_city_number_raises_error(self):
        board = Board(city_count=5, epicenter=3)

        with self.assertRaises(IndexError):
            board.move(0)

        with self.assertRaises(IndexError):
            board.move(6)


class SolverTest(unittest.TestCase):
    def test_solver_finds_middle_epicenter(self):
        board = Board(city_count=7, epicenter=4)
        solver = Solver(board)

        self.assertEqual(solver.solve(), 4)

    def test_solver_finds_left_edge_epicenter(self):
        board = Board(city_count=7, epicenter=1)
        solver = Solver(board)

        self.assertEqual(solver.solve(), 1)

    def test_solver_finds_right_edge_epicenter(self):
        board = Board(city_count=7, epicenter=7)
        solver = Solver(board)

        self.assertEqual(solver.solve(), 7)

    def test_solver_uses_no_arguments(self):
        board = Board(city_count=3, epicenter=2)
        solver = Solver(board)

        self.assertEqual(solver.solve(), 2)

    def test_solver_works_for_many_board_sizes(self):
        for size in range(1, 15):
            for epicenter in range(1, size + 1):
                with self.subTest(size=size, epicenter=epicenter):
                    board = Board(city_count=size, epicenter=epicenter)
                    solver = Solver(board)

                    self.assertEqual(solver.solve(), epicenter)


if __name__ == "__main__":
    unittest.main(argv=[""], exit=False)

board = Board(12, 9)
solver = Solver(board)

print(board.move(3))

print(solver.solve())
```
