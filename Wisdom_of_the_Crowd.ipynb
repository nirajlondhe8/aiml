```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters of the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Define sample sizes to examine the evolution
sample_sizes = np.arange(10, 1001, 10)

# Arrays to store the sample means and variances
means = []
variances = []

# Generate samples and calculate the statistics
for size in sample_sizes:
    # Generate a sample of the given size from the normal distribution
    sample = norm.rvs(loc=mu, scale=sigma, size=size)

    # Compute the sample mean and variance
    sample_mean = np.mean(sample)
    sample_variance = np.var(sample, ddof=1)  # Use ddof=1 for unbiased variance

    # Store the results
    means.append(sample_mean)
    variances.append(sample_variance)

# Plot the evolution of the sample mean
plt.figure(figsize=(12, 5))

# Plot the mean
plt.subplot(1, 2, 1)
plt.plot(sample_sizes, means, label='Sample Mean', color='blue')
plt.axhline(y=mu, color='red', linestyle='--', label='True Mean (μ)')
plt.xlabel('Sample Size')
plt.ylabel('Mean')
plt.title('Convergence of Sample Mean to True Mean')
plt.legend()

# Plot the variance
plt.subplot(1, 2, 2)
plt.plot(sample_sizes, variances, label='Sample Variance', color='green')
plt.axhline(y=sigma**2, color='red', linestyle='--', label='True Variance (σ²)')
plt.xlabel('Sample Size')
plt.ylabel('Variance')
plt.title('Convergence of Sample Variance to True Variance')
plt.legend()

plt.tight_layout()
plt.show()
```
