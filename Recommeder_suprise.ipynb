```python
from google.colab import files
import os
import shutil
import pandas as pd

# Step 1: Upload CSV file
uploaded = files.upload()

# Step 2: Create 'data' directory
if not os.path.exists('data'):
    os.makedirs('data')

# Step 3: Move the uploaded file to 'data' directory
for filename in uploaded.keys():
    shutil.move(filename, os.path.join('data', filename))

# Step 4: Verify the file is in 'data' directory
print("Files in 'data' directory:", os.listdir('data'))

# Step 5: Read the CSV file
file_path = 'data/movie_ratings.csv'  # Replace 'your_file.csv' with the actual file name
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
```

```python

# Step 1: Install the Surprise Library
!pip install scikit-surprise

# Step 2: Import Necessary Libraries
from surprise import Dataset, Reader, SVD
import pandas as pd
from surprise.model_selection import cross_validate

# Step 3: Load the Data
# Load the movie ratings data into a pandas DataFrame
movie_ratings = pd.read_csv('data/movie_ratings.csv', index_col=0)
print(movie_ratings.head())

# Step 4: Problem 1 - Loading the Dataset
# Select relevant columns for Surprise
a = movie_ratings[['userId', 'title', 'rating']]

# Define a reader with the rating scale from 0 to 5
reader = Reader(rating_scale=(0, 5))

# Load the dataset from the DataFrame
sf = Dataset.load_from_df(a, reader)

# Build the full trainset
train = sf.build_full_trainset()

# Check the types of the created objects
print(type(sf))
print(type(train))

# Step 5: Problem 2 - Instantiate the SVD Model
# Instantiate the SVD model with 2 latent factors
model = SVD(n_factors=2)

# Check the number of factors in the model
print(model.n_factors)

# Step 6: Problem 3 - Fitting the Model
# Fit the SVD model to the training data
model.fit(train)

# Verify the model
print(model)

# Step 7: Problem 4 - Making Predictions
# Build a testset from the train set
test = train.build_testset()

# Predict ratings for the testset
predictions_list = model.test(test)

# Display the first few predictions
print(predictions_list[:5])

# Step 8: Problem 5 - Cross-Validate the Model
# Cross-validate the model using RMSE and 5 folds
cross_val_results = cross_validate(model, sf, measures=['RMSE'], cv=5, verbose=True)

# Print the cross-validation results
print(cross_val_results)
```
